```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
death_cause = pd.read_csv('data/Death Cause Reason by Country.csv')
```

```python
death_cause.drop(columns= 'Unnamed: 32', axis= 1, inplace= True)
```

```python
death_cause.isna().sum()
```

```python
death_cause.describe()
```

```python
# 정렬
```

```python
# 나라이름으로 데이터프레임 정렬
death_cause.sort_values('Country Name', ascending= True)
```

```python
# 총 사망자 수로 정렬
```

```python
death_cause.sort_values('total of death', ascending= False)
```

```python
# 전세계에서 어떤 질병때문에 가장 많이 사망 하였는지
```

```python
death_cause.iloc[:,1:].sum().sort_values().plot(kind= 'barh')
plt.title('which diseases are the most common')
plt.xlabel('Count of Death')
plt.show()
```

```python
# 전세계에서 어떤 국가에 가장 많은 사망자가 있는지
```

```python
death_cause.iloc[:,1:]
```

```python
death_cause['total of death'] = death_cause.sum(axis=1, numeric_only=True)
```

```python
plt.figure(figsize=(10,20))
death_cause[['Country Name', 'total of death']].plot()
plt.title('Total number of deaths by country')
plt.xlabel('Country')
plt.ylabel('Total of Death')
plt.show()
```

```python
# 내가 원하는 국가를 입력하면 국가별 사망자 수와 국가별로 어떤 질병이 가장 많이 발생하는지에 대해서 나타냄
```

```python
# 포함되는 단어를 입력하면, 그 단어를 선택해서 
```

```python
death_cause[death_cause['Country Name'].str.contains('south', case= False)]['Country Name']
```

```python
death_cause[death_cause['Country Name'] == 'South Korea']
```

```python
# 선택한 국가의 사망자 수 
death_cause[death_cause['Country Name'] == 'South Korea'][['Country Name', 'total of death']]
```

```python
# 어떤 원인이 가장 많이 발생하는지
```

```python
data = death_cause[death_cause['Country Name'] == 'South Korea'].iloc[:,0:-1]
```

```python
data = data.set_index('Country Name')
```

```python
# 상위 10개의 사망 원인을 bar 그래프로
```

```python
data_top10 = data.sum().sort_values(ascending= False).head(15)
```

```python
data_top10
```

```python
data_top10.plot(kind= 'bar')
plt.title('Cause of death in ' + data.index.values[0] + ' Top 15')
plt.show()
```
